# 用纯Python代码对时区进行计数

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]  # 不是所有记录中都有时区，故需要加if判断条件
time_zones[:10]
```

```python
# 只使用标准python库对时区进行计数
# 方法一
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts

# 方法二
from collections import defaultdict
def get_counts_2(sequence):
    counts = defaultdict(int) # 所有值均初始化为0
    for x in sequence:
        counts[x] += 1
    return counts

counts  = get_counts_2(time_zones)
counts['America/New_York']
```

```python
# 获取前10为的时区及其计数值
# 方法一
def top_counts(count_dict,n=10):
    value_key_pairs = [(count,tz) for tz,count in count_dict.items()]
    value_key_pairs.sort() # 按count进行升序排序
    return value_key_pairs[-n:]
top_counts(counts)
```

```python
# 方法二
# collections.Counter类 可以计数
from collections import Counter
counts = Counter(time_zones)
counts.most_common(10)  # 获取top10的时区及其计数值
```

# 用pandas对时区进行计数

```python
import pandas as pd 
import numpy as np
frame = pd.DataFrame(records)
frame.head()
```

```python
frame['tz'][:10]
```

```python
tz_counts = frame['tz'].value_counts()  #value_counts()计数  并按降序排列
tz_counts[:10]
```

```python
# 时区中未知或缺失时区填上替代值
# fillna()替换缺失值（NA），未知值（空字符串）可通过布尔型数组索引加以替换
clean_tz = frame['tz'].fillna('Missing')
clean_tz[clean_tz == ''] = 'Unknown'
tz_counts = clean_tz.value_counts()
tz_counts[:10]
```

```python
# 绘制水平条形图
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.figure()
tz_counts[:10].plot(kind = 'barh',rot = 0)
```

```python
# a字段含有执行URL短缩操作的浏览器、设备、应用程序的相关信息
frame['a'][1]
frame['a'][50]
frame['a'][51]
```

```python
# 分离出浏览器相关信息
results = pd.Series([x.split()[0] for x in frame.a.dropna()])
results[:5]
```

```python
results.value_counts()[:8]
```

```python
# 按Windows 和 非Windows 用户对时区统计信息进行分解
# 移除a字段中的缺失值
# a中字符串包含‘Windows’，则为 Windows用户
cframe = frame[frame.a.notnull()]
operating_system = np.where(cframe['a'].str.contains('Windows'),'Windows','非Windows')
operating_system[:5]
```

```python
# 根据时区和新的操作系统列表对数据进行分组
# size（）对分组结果进行计数，相当于上面的value_counts函数
# unstack()对计数结果进行重塑,不堆积，即将操作系统变成行索引；stack() 堆积，将操作系统变成第二列索引
by_tz_os = cframe.groupby(['tz',operating_system])
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
# 选取最常出现的时区
# 用于按升序排序
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
# take 按照这个顺序截取租后10行
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
# 绘制堆积条形图
count_subset.plot(kind='barh',stacked =True)
plt.title('按Windows和非Windows用户统计的最常出现的时区')
```

```python
normed_subset = count_subset.div(count_subset.sum(1),axis = 0)
normed_subset.plot(kind='barh',stacked =True)
plt.title('按Windows和非Windows用户比例统计的最常出现的时区')
```

# 1880-2010年间的全美婴儿姓名

```python
import pandas as pd
# 由于文件名中带有中文，read_csv会报错，所以，先open，再传入f 就不会报错
f = open(prefix+'/datasets/babynames/yob1880.txt')
names1880 = pd.read_csv(f,names=['name','sex','births'])
names1880
```

```python
names1880.groupby('sex').births.sum()
```

```python
# 所有文件中仅含有当年出现超过5词的名字
# 将所有数据组装到一个DataFrame中，并加上year字段
# 2010是目前最后一个有效统计年份
years = range(1880,2011)
pieces = []
columns =['name','sex','births']

for year in years:
    path = prefix +'/datasets/babynames/yob'+ str(year) + '.txt'
    f= open(path)
    frame = pd.read_csv(f,names=columns)
    
    frame['year']=year
    pieces.append(frame)

# 将所有护具整合到单个DataFrame中
names = pd.concat(pieces,ignore_index = True)  # ignore_index = True 不保留read_csv所返回的原始行号
names
```

```python
total_births = names.pivot_table('births',index = 'year',columns='sex',aggfunc='sum')
total_births.tail()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.figure()
total_births.plot(title = '按性别和年度统计总出生数')
plt.show()
```

```python
# 增加prop列，用于存放指定名字的婴儿数相对于总出生数的比例
def add_prop(group):
    # 整数除法会想下取整
    births = group.births.astype(float)
    
    group['prop'] = births /births.sum()
    return group 
# 先按year 和 sex进行分组，再将新列加到各个分组上
names = names.groupby(['year','sex']).apply(add_prop)
names
```

```python
# 验证所有分组的prop的综合是否为1
import numpy as np
np.allclose(names.groupby(['year','sex']).prop.sum(),1)
```

```python
# 获取每对year/sex组合的前1000个名字
# 方法一
def get_top1000(group):
    return group.sort_index(by = 'births',ascending=False)[:1000]
grouped = names.groupby(['year','sex'])
top1000 = grouped.apply(get_top1000)
top1000
```

```python
# 方法二
pieces =[]
for year ,group in names.groupby(['year','sex']):
    pieces.append(group.sort_index(by = 'births',ascending=False)[:1000])
top1000 = pd.concat(pieces,ignore_index=True)
top1000
```

# 分析命名趋势

```python
# 将前1000个名字分为男女两部分
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']

# 生成一张按year和name统计的总出生透视表
total_births = top1000.pivot_table('births',index = 'year',columns = 'name',aggfunc = 'sum')

total_births
```

```python
# 绘制几个名字的曲线图
subset=total_births[['John','Harry','Mary','Marilyn']]
subset.plot(subplots=True,figsize=(12,10),grid = False,title='Number of births per year')
```

```python
# 评估命名多样性的增长
#计算最流行的1000个名字所占的比例
table = top1000.pivot_table('prop',index='year',columns='sex',aggfunc='sum')
table.plot(title='Sum of table1000.prop by year and sex',yticks=np.linspace(0,1.2,13),xticks=range(1880,2020,10))
```

```python
# 计算占总出生人数前50%的不同名字的数量
# 仅考虑2010年男孩的名字
df = boys[boys.year == 2010]
df
```

```python
# 对prop降序排雷，获取前面多少个名字加起来才能够到50%
prop_cumsum = df.sort_index(by='prop',ascending=False).prop.cumsum()
prop_cumsum[:10]
```

```python
# searchsorted 可以找出0.5应该插在哪个位置才能保证不破坏顺序
prop_cumsum.searchsorted(0.5)  # 结果为116，但是由于索引是从0开始，因此这个结果要加1
```

```python
df =boys[boys.year == 1900]
in1900 = df.sort_index(by='prop',ascending=False).prop.cumsum()
in1900.searchsorted(0.5)+1
```

```python
# 对所有year/sex组合计算各分组的这个值了

def get_quantile_count(group,q=0.5):
    group = group.sort_index(by='prop',ascending=False)
    return group.prop.cumsum().searchsorted(q)[0]+1

diversity = top1000.groupby(['year','sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
diversity.head()
```

```python
diversity.plot(title='Number of popular names in top 50%')  # 图中可以看出女孩名字的多样性远高于男孩
```

# 最后一个字母的变革

```python
# 从name列取出最后一个字母
get_last_letter = lambda x:x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letter'

table = names.pivot_table('births',index = last_letters,columns=['sex','year'],aggfunc='sum')
# 选出具有代表性的三年
subtable = table.reindex(columns=[1910,1960,2010],level='year')
subtable.head()
```

```python
subtable.sum()
```

```python
# 各性别各末尾字母占总出生人数的比例
letter_prop = subtable/subtable.sum().astype(float)
letter_prop
```

```python
import matplotlib.pyplot as plt
fig,axes = plt.subplots(2,1,figsize=(10,8))
letter_prop['M'].plot(kind='bar',rot=0,ax=axes[0],title='Male')
letter_prop['F'].plot(kind='bar',rot=0,ax=axes[1],title = 'Female',legend=False)
```

```python
# 在男孩名字中选取几个字母做时间序列
letter_prop = table/table.sum().astype(float)
dny_ts = letter_prop.ix[['d','n','y'],'M'].T
dny_ts.head()
```

```python
dny_ts.plot(title='各年出生的男孩中名字以d/n/y结尾的人数比例')
```

## 变成女孩名字的男孩名字

```python
#在top1000找到以‘lesl’开头的一组名字
all_names = top1000.name.unique()
mask = np.array(['lesl' in x.lower() for x in all_names])
lesley_like = all_names[mask]
lesley_like
```

```python
# 过滤掉其他名字后，按名字分组计算出生数
filtered = top1000[top1000.name.isin(lesley_like)]
filtered.groupby('name').births.sum()
```

```python
# 按性别和年度进行聚合
table = filtered.pivot_table('births',index='year',columns='sex',aggfunc='sum')
table = table.div(table.sum(1),axis=0)
table.tail()
```

```python
table.plot(title ='各年度使用‘Lesley型’名字的男女比例',style={'M':'k-','F':'k--'})
```
